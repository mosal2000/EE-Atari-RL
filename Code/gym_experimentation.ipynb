```python
import gym
from gym.utils import play
import matplotlib.pyplot as plt
import time
```

```python
env_name = 'ALE/SpaceInvaders-ram'
env = gym.make(env_name,render_mode="human")
```

```python
env.reset()
for step in range(20):
    random_action = env.action_space.sample()
    observation,reward,terminated, turnicated,info = env.step(random_action)
    
    print(f"Observation: {observation}")
    print(f"Reward: {reward}")
    print(f"Done: {terminated},{turnicated}")
    print(f"Info: {info}")

env.close()
```
